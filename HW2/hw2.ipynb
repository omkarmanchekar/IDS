```python
import math


less_than_50=len(train_dataset.loc[train_dataset['class']=='<=50K'])
greater_than_50=len(train_dataset.loc[train_dataset['class']=='>50K'])

less_than_50_prob = less_than_50/(greater_than_50+less_than_50)
greater_than_50_prob = greater_than_50/(greater_than_50+less_than_50)


mean = train_dataset.groupby(['class']).mean()
std_var = train_dataset.groupby(['class']).std()

z=train_dataset.groupby(['class']).count()


def fit_normal_distribution(column ,value,class_value):
    if class_value== '<=50K':
        i = 0
    if class_value == '>50K':
        i = 1
    x= 1/(std_var[column][i]*(math.sqrt(2*math.pi)))
    y = pow((value - mean[column][i])/std_var[column][i],2)
    normal=x * math.exp(-0.5*y)
    return normal

def probability(column,value,class_value):
    if class_value== '<=50K':
        denominator = less_than_50
    if class_value == '>50K':
        denominator = greater_than_50
    f=train_dataset.loc[train_dataset['class']==class_value].loc[train_dataset[column] == value]
    Nic= len(f)
    Nc= denominator
    laplace = (Nic +10)/(Nc + 2)
    return laplace





    
```

```python
classification = []
for index,row in test_dataset.iterrows():
    less_than_prob_percentage = less_than_50_prob * probability('sex',row['sex'],'<=50K')  * probability('race',row['race'],'<=50K')  * probability('relationship',row['relationship'],'<=50K')  * probability('occupation',row['occupation'],'<=50K')  * probability('marital-status',row['marital-status'],'<=50K')  * probability('education',row['education'],'<=50K')  * probability('workclass',row['workclass'],'<=50K') * fit_normal_distribution('age',row['age'],'<=50K') * fit_normal_distribution('fnlwgt',row['fnlwgt'],'<=50K') * fit_normal_distribution('education-num',row['education-num'],'<=50K') * fit_normal_distribution('hoursperweek',row['hoursperweek'],'<=50K')
    greater_than_prob_percentage = greater_than_50_prob * probability('sex',row['sex'],'>50K')  * probability('race',row['race'],'>50K')  * probability('relationship',row['relationship'],'>50K')  * probability('occupation',row['occupation'],'>50K')  * probability('marital-status',row['marital-status'],'>50K')  * probability('education',row['education'],'>50K')  * probability('workclass',row['workclass'],'>50K') * fit_normal_distribution('age',row['age'],'>50K') * fit_normal_distribution('fnlwgt',row['fnlwgt'],'>50K') * fit_normal_distribution('education-num',row['education-num'],'>50K') * fit_normal_distribution('hoursperweek',row['hoursperweek'],'>50K')
    print(less_than_prob_percentage)
    print(greater_than_prob_percentage)
    if less_than_prob_percentage > greater_than_prob_percentage:
        classification.append("<=50K")
    if less_than_prob_percentage < greater_than_prob_percentage:
        classification.append(">50K")
 

```
