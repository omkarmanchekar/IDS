<h2>LOAD DATA</h2>

```python
import pandas as pd
import numpy as np
dataset = pd.read_csv('https://clo-pfw-prod.s3.us-west-2.amazonaws.com/data/PFW_2021_public.csv')
dataset
```

<h2>REMOVE REDUNDANCY AND USELESS ATTRIBUTES</h2>

```python
dataset=dataset.drop_duplicates('loc_id').dropna(how='any').reset_index(drop=True)
dataset=dataset.drop(['loc_id','longitude','PROJ_PERIOD_ID','species_code','reviewed','latitude','entry_technique','sub_id','obs_id','valid','day1_am','day1_pm','day2_am','day2_pm','Data_Entry_Method','effort_hrs_atleast'],axis=1)
# np.unique(dataset[''])
dataset

```

<H2>COMBINE DAY,MONTH AND YEAR ATTRIBUTE AS DATE</H2>

```python
combine_datetime=pd.to_datetime(dataset[['Day','Month','Year']])
dataset['Date']= combine_datetime
dataset=dataset.drop(['Month','Day','Year'],axis=1)
dataset
```

<h2>CATEGORICAL DATA</h2>

```python
cat={
    0.0000e+00:'Sunny',1.0000e-03:'Sunny',5.0000e+00:'Snow',1.5001e+01:'Snow Storm'
}
dataset['snow_dep_categorical_label']=dataset['snow_dep_atleast'].map(cat)
dataset
sub_list={}
x=0
for i in np.unique(dataset['subnational1_code']):
    sub_list[i]=x
    x=x+1
sub_list
dataset['subnational1_code_label']=dataset['subnational1_code'].map(sub_list)
dataset
```

<H2>BINNING ATTRIBUTE 'HOW_MANY' INTO SIZE OF 10</H2>

```python
bins=np.arange(0,501,10)
labels = ['{} - {}'.format(i + 1, j) for i, j in zip(bins[:-1], bins[1:])] 
labels[0] = '0 - 10'

bin_array=pd.cut(dataset['how_many'],bins,labels=labels,right=True,include_lowest=True)
dataset['how_many_bins']=bin_array
dataset
```
