```python
import mmh3
import math
```

```python
input=[]
input_set={}
fp=0
tn=0
file = open("/Users/omkarmanchekar/Desktop/Semester 1/Intro to Data Science/HW/HW6/Dataset_Q3/listed_username_30.txt", "r", encoding='ISO-8859-1')
for line in file:
	input.append(line)
	input_set[line]=line
file.close()
```

```python
n = len(input)
m = 4 * n
bit_vector = [0]*m
print("k =",math.ceil((m/n)*math.log(2)))
```

```python
for i in input:
	h1 = mmh3.hash(i,20) % m
	h2 = mmh3.hash(i,35) % m
	h3 = mmh3.hash(i,49) % m
	bit_vector[h1] = bool(1)
	bit_vector[h2] = bool(1)
	bit_vector[h3] = bool(1)
```

```python
file = open("/Users/omkarmanchekar/Desktop/Semester 1/Intro to Data Science/HW/HW6/Dataset_Q3/listed_username_365.txt", "r",encoding='ISO-8859-1')
for line in file:
	h1 = mmh3.hash(line, 20) % m
	h2 = mmh3.hash(line, 35) % m
	h3 = mmh3.hash(line, 49) % m
	if (bit_vector[h1] is bool(1)) and (bit_vector[h2] is bool(1)) and (bit_vector[h3] is bool(1)):
		if line not in input_set:
			fp = fp + 1
	else:
		tn = tn + 1
file.close()
```

```python
print (((fp)/(fp + tn)) * 100, "%")
```
